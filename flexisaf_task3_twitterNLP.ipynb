```python
!pip install nltk emoji pyspellchecker

import re
import string
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from emoji import demojize
from spellchecker import SpellChecker
```

```python
# finding emoticons

emoticon_regex = r"[:;=8][-~]?[)DdPp(/\(\|\\<>]"

def find_emoticons(text):
    return re.findall(emoticon_regex, str(text))

all_found = []
for tweet in df_sample['text']:
    all_found.extend(find_emoticons(tweet))

emoticon_counts = Counter(all_found)

for emo, count in emoticon_counts.most_common(20):
    print(f"{emo} : used {count} times")
```

```python
#EMOTICONS

EMOTICON_DICT = {
    # Skeptical / Annoyed
    ":/": "skeptical", "=/": "skeptical", ":-/": "skeptical", ";/": "skeptical",

    # Happy / Laughing
    "=d": "laughing", ";d": "laughing", ":-d": "laughing", "8d": "laughing",

    # Wink / Joke
    ";)": "wink", ";-)": "wink",

    # Playful / Tongue Out
    ";p": "playful", ":p": "playful", "=p": "playful", ":-p": "playful", "8p": "playful",

    # Neutral / Bored
    ":|": "neutral",

    # Sad
    ";(": "sad", "=(": "sad"
}
```

```python
#finding URLs

url_regex = r'https?://\S+|www\.\S+|\S+\.(?:com|net|org|edu|gov|ly|bit|gl|co)\S*'

def find_url(text):
  return re.findall(url_regex, str(text))

all_found = []
for i in df_sample["text"]:
    all_found.extend(find_url(i))

url_count = Counter(all_found)

for url, count in url_count.most_common(100):
    print(f"{url} used {count} times")



```

```python
url_email_regex = r'(\S+@\S+\.\S+)|(https?://\S+|www\.\S+|\S+\.(?:com|net|org|edu|gov|ly|bit|gl|co|me|tv|us)\S*)'

def clean(text):
  text = str(text).lower()

  #Removing emails and url
  text = re.sub(url_email_regex, ' ', text)

  #Replacing emoticons with word
  for emo, word in EMOTICON_DICT.items():
    text = text.replace(emo, f"{word}")

  #Replacing emojis with words
  text = demojize(text).replace(':',' ').replace('_', ' ')

  #Removal of punctuation
  text = text.translate(str.maketrans('', '', string.punctuation))

  #Splitting text into words
  words = text.split()

# (Speed optimization: only check words not in dictionary
  misspelled = spell.unknown(words)
  words = [spell.correction(w) if w in misspelled else w for w in words]
  words = [w for w in words if w is not None] # Remove empty results

  # Removal of Frequent words (Stopwords)
  words = [w for w in words if w not in stop_words]

  # Lemmatization (Base dictionary form)
  words = [lemmatizer.lemmatize(w) for w in words]

  return " ".join(words)
```

```python
# Removal of Rare and Dataset-wide Frequent words
def remove_extremes(series, freq_limit=100, rare_limit=1):
    all_words = " ".join(series).split()
    counts = Counter(all_words)

    # Identify words that appear too much or only once
    to_remove = set([w for w, c in counts.items() if c > freq_limit or c <= rare_limit])

    return series.apply(lambda x: " ".join([w for w in x.split() if w not in to_remove]))
```

```python
from tqdm import tqdm
tqdm.pandas()

df_sample['cleaned_text'] = df_sample['text'].progress_apply(clean)

df_sample['cleaned_text'] = remove_extremes(df_sample['cleaned_text'])

print(df_sample[['text', 'cleaned_text']].head(10))
```
